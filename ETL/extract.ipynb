```python
VAR_LIST_PATH = Path('docs/vars_list_with_descriptions.txt')
```

```python
NEW_VAR_NAMES = [
    "SurveyDate",
    "State",
    "Sex",
    "GeneralHealth",
    "PhysicalHealthDays",
    "MentalHealthDays",
    "LastCheckupTime",
    "PhysicalActivities",
    "SleepHours",
    "RemovedTeeth",
    "HadHeartAttack",
    "HadAngina",
    "HadStroke",
    "HadAsthma",
    "HadSkinCancer",
    "HadCOPD",
    "HadDepressiveDisorder",
    "HadKidneyDisease",
    "HadArthritis",
    "HadDiabetes",
    "DeafOrHardOfHearing",
    "BlindOrVisionDifficulty",
    "DifficultyConcentrating",
    "DifficultyWalking",
    "DifficultyDressingBathing",
    "DifficultyErrands",
    "SmokerStatus",
    "ECigaretteUsage",
    "ChestScan",
    "RaceEthnicityCategory",
    "AgeCategory",
    "HeightInMeters",
    "WeightInKilograms",
    "BMI",
    "AlcoholDrinkers",
    "HIVTesting",
    "FluVaxLast12",
    "PneumoVaxEver",
    "TetanusLast10Tdap",
    "HighRiskLastYear",
    "CovidPos"
]
```

```python
var_list_df = pd.read_csv(VAR_LIST_PATH, sep=' - ', header=None, names=['Variable', 'Description'], engine='python')
```

```python
var_list_df
```

```python
var_list = var_list_df['Variable'].to_numpy()
var_list
```

```python
heart_df = heart_df[var_list]
```

```python
heart_df.columns = NEW_VAR_NAMES
```

```python
heart_df.insert(0, 'id', heart_df.index + 1)
```

## Loading the data into staging 1

```python
from dotenv import load_dotenv
import os

load_dotenv()

user = os.getenv('DB_USER')
password = os.getenv('DB_PASSWORD')
host = os.getenv('DB_HOST')
port = os.getenv('DB_PORT')
db_staging = os.getenv('DB_STAGING')
```

```python
engine = create_engine(f'mysql+pymysql://{user}:{password}@{host}:{port}/{db_staging}')
```

```python
heart_df.to_sql("heart_data", con = engine, if_exists= 'replace', index= False)
```
